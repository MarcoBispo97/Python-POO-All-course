# Dice Game project:

# Aggregation:
- Concept in object-oriented programming that describes the `relationship` between two class
- Building complex onjects that are `composed` of other objects.
- Class B can use the functionality of Class B
- Instance A ---> need instance B
- We `store instances` of the simpler objects in the more complex objetects
- It's `has a` relationship between the classes. An object of class B has an object of class A.
- Example: Employye `has a` vehicle
- Work with the vehicle inside the Employee class

## Agregation example

```python
class Vehicle:
    def __init__(self, color, license_plate, is_electric=False):
        self.color = color
        self.license_plate = license_plate
        self.is_electric = is_electric

    def show_license_plate(self):
        print(f"License Plate: {self.license_plate}")
    
    def show_info(self):
        info = f"Color: {self.color}, License Plate: {self.license_plate}, Electric: {self.is_electric}"
        return info
    

class Employee:
    def __init__(self, name, vehicle: Vehicle):
        self.name = name
        self.vehicle = vehicle

    def show_vehicle_info(self):
        print(f"Employee Name: {self.name}")
        # Here it is very important to delegate the call to the Vehicle instance
        return self.vehicle.show_info()

# Example usage
vehicle1 = Vehicle(color="Red", license_plate="ABC123", is_electric=True)
employee1 = Employee(name="John Doe", vehicle=vehicle1)
employee1.show_vehicle_info()
print(employee1.vehicle.show_info())
print(employee1.vehicle)

vehicle2 = Vehicle(color="Blue", license_plate="XYZ789", is_electric=False)
employee2 = Employee(name="Jane Smith", vehicle=vehicle2)
employee2.show_vehicle_info()
print(employee2.vehicle.show_info())
print(employee2.vehicle)
```
