# Instance Attributes: __init__()
- `__init__()`:
    - Special method used to define the initial state of an object.
    - Called automatically when an instance is created
``` python

class House:
    # __init__: The constructor. Called automatically when a new object is created.
    # recive the input attributes
    # Its purpose is to initialize the object's state.
    def __init__(self, price):

        # self: A reference to the current instance of the class (the object itself).
        # Used to access or set attributes on the object.

        # self.price: An "instance attribute".
        # This attaches the 'price' to the specific object ('self').
        self.price = price

# --- Example ---
# house_a = House(500000)
# 1. A new House object is created.
# 2. __init__ is called on this new object.
# 3. The object is passed as 'self'; 500000 is passed as 'price'.
# 4. 'self.price = 500000' creates the price attribute on house_a.
```

``` python
class Carro:
    # Aqui, marca, modelo e ano são parâmetros de entrada
```

# Instances and `self` OOP:
- Classes define the state and the gehavior of the objects in a `generic` way
- It is not specific. The code in the class has to work for `any` instance of the class
- `self` is a generic way of refering to the current instance of the class
``` python
class House:
    def __init__(self, price):
        # 1 Assign the value of the variable price
        # 2 To the instance attribute "price"
        # 3 Of the instance that is being created
        # price is the parameter of the init method
        # self.price is the attribute of the instance
        # These elements are not equal
        self.price = price
```

# How to create Instances in Python:
-  `self` is "skipped" in the arguments list when create an instance
    - No arguments:
        <ClassName>()
        ``` python
        class Backpack:
            def __init__(self):
                self.items = []
        
        my_backpack = Backpack()
        print(my_backpack)
        # Return
        # <__main__.Backpack object at 0x0000029A2DD11010>
        # memory location
        ```
    - Whith arguments:
        <ClassName>(<value>)
        ```` python
        class Circle:
            def __init__(self, radius):
                self.radius = radius
        radius = 3
        my_circle = Circle(radius)
        print(my_circle)
        # Return
        # <__main__.Backpack object at 0x0000029A2DD11010>
        # memory location
        ````
    - Multiple arguments:
        <ClassName>(<arguments>)
        ``` python
        class Rectangle:
            def __init__(self, length, width):
                self.length = length
                self.width = width

        my_rectangle = Rectangle(3, 6) # 2 arguments here !
        # Return
        # <__main__.Rectangle object at 0x0000029A2DD11550>
        # memory location
        ```




`````python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

my_rectangle = Rectangle(3, 6)
print(my_rectangle)
`````

# Default Arguments
- Default values assigned to parameters when their correspomnding arguments are omitted
-`<parameter>=<value>`
- Correct:
    ```python
    def complex(real, imag=0.0):
        return magic(r=real, i-imag)
    ```
- Wrong:
    ``` python
    def complex(real, imag = 0.0):
        return magic(r = real, i = imag)
    ```
- Example:
    ```python
    class Circle:
        def __init__(self, radius=5):
            self.radius = radius
    my_circle = Circle()
    print(my_circle.radius)
    # Return 5
    my_circle = Circle(10)
    print(my_circle.radius)
    ```


`````python
class Circle:
    def __init__(self, radius=5):
        self.radius = radius
        self.color = "Blue"
my_circle = Circle()
print(my_circle.radius)
# Return 5
my_circle = Circle(10)
print(my_circle.radius)
print(my_circle.color)
`````
