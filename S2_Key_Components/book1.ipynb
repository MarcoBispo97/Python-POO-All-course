# OOP Object-Oriented Programming:
- A `programming paradigm` that organizes software design around objects.
- **Programming Paradigm**: an `approach used to solve problems with code.`
- A style or 'way' of programmming.
- OOP is based on `messages` sento to objects
- Objects **respond** to these messages by performing actions.
- Objects have their own **state** or set of attributes.
- Programming paradigms are **not** mutually exclusive (diferentes paradigmas de programação (como orientado a objetos, funcional, procedural, etc.) podem ser usados juntos no mesmo projeto ou até mesmo no mesmo código. Eles não se excluem — você pode combinar técnicas de vários paradigmas para resolver problemas de forma mais eficiente, aproveitando o que cada abordagem tem de melhor. Por exemplo, em Python, é comum misturar programação orientada a objetos com programação funcional).
- The same program can have **multiple** programming paradigms.

# Advantages of Object-Oriented Programming
- `Modularity`: Mostly independent each other, easly to change without afect all the program

- `Extensibility`: Combine to create funcionality, objects can be **extended** to include new attributes and behavior.

- `Reusability`: Objects can be **reused** within the program and in future projects, fasters development and lower cost.

- State + Behavior = New functionality very easy.

- Easy to bug fix.

# Important acronyms:

- `OOP` stands for "Object Oriented Programming."

- `PEP` stands for "Python Enhancement Proposal." It's a document that outlines new Python features and processes. PEP 8 established the official Python style guide.

- `OOAD` stands for "Object Oriented Analysis and Design." It's an approach that models systems as a collection of objects that interact with each other.
