# Wildcard Import Statements
``` python
from <module> import *
```
``` python

print(cos) 
```

````python
from math import *
print(pi)
print(sin)
print(cos) 
````

- Import All the Elements of a Module:
    - You can use this syntax to import all the elements of a module
        - `import <module> as <new_name>`
    - For example:
        - `import random as r`
    - If you use this syntax to import the module, you will need to add the name of the module followed by a dot and the name of the element:
        - `<new_name>.<element>`

- Import Specific Elements:
    - You can use this syntax to import specific elements from a module with a new name:
        - `from <module> import <element> as <new_name>`
        - To access the element, you would write the new name of the element::
            `<new_name>`

- Absolute Imports:
    - They have full path to your script (python file), starting from the root folder
    - The folders are separated with a period
    - They can be as long as needed
    - `Advantage: ` They are the preferred type of import statement because they clearly show the path to the module.
    - `Disadvantage: ` Yhey can be quite long if the project has many subfolders
    - Example:
        ``` python
        from <package>.<module> import <module>
 
        import <package>.<module>.<function>
        ```
    - Tip: <package>, <module>, <function> are placeholders for the real names. You need to replace them in the project.

- Relative Imports:
    - The path specifies where the modules are located relative to the current  file.
    - The syntax uses leading dots: one dot indicates the directory of the current script. Two dots indicate the parent directory, and so on, one directory per dot.
    - The dots can only go up to the directory of the current file.
    - Example:
        ``` python
        import other
 
        from .<module> import <element>
        
        from ..<subpackage>.<module> import <element>
        ```
