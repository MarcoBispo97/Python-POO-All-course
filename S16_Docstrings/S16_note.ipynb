# Docstring in Python
- Documentation + strings = Docstring
- PEP 257: -- Docstring Conventions
    - A docstring is a string literal thta occours as the first statement in a module, function, class or method definition. Such a doc-string becomes the `__doc__` special attribute of that object.
    - For consistency, always use """triple double quotes""" around docstrings 
    - Humman-readable comments that explain purpose and logic.
- Content:
    - Purpose and Logic
    - Arguments (description, data types)
    - Return value
    - Relevant information:
        - Valid range
        - Exceptions
- Docstrings are the `first` statement in a class, method, module, or function. 
- They sould be included in all `public` methods, including `__init__()`

# Docstring vs Comments:
- Docstrings: 
    - They are "linked" to the element they describe
    - `Can` be used to generate documentation automatically.
    - Cabn be read with `help()` in the interactive shell.
- Comments:
    - They are `not` linked to any part of the code.
    - `Can't` be used to generate documentation automatically.
    - Can only be read `directly` in the source code.

````python
class Flight:
    """A class that represents a commercial airline flight.

    This class manages flight information including the flight number, route details,
    passenger count, and crew information. It provides properties to access and modify
    sensitive flight data with validation.

    Attributes:
        number (str): The unique flight identifier (e.g., "AA123")
        origin (str): The departure airport code
        destination (str): The arrival airport code
        num_passengers (int): The number of passengers on the flight
        total_weight (float): The total weight of the aircraft with cargo and passengers
            (in pounds, must be positive and a float)
        pilot (str): The name of the pilot in command
        crew (list): A list of crew member names

    Methods:
        display_flight_data(self):
            Display all flight information in a formatted manner
    """

    def __init__(
        self, number, origin, destination, num_passengers, total_weight, pilot, crew
    ):
        """Initialize an instance of Flight.

        Creates a new Flight object with the provided flight details including route
        information, passenger count, weight, and crew assignment.

        Arguments:
            number (str): The unique flight identifier (e.g., "AA123")
            origin (str): The departure airport code (e.g., "JFK")
            destination (str): The arrival airport code (e.g., "LAX")
            num_passengers (int): The number of passengers on the flight
            total_weight (float): The total weight of the aircraft with cargo and
                passengers (must be positive)
            pilot (str): The name of the pilot in command
            crew (list): A list of crew member names
        """
        self.number = number
        self.origin = origin
        self.destination = destination
        self.num_passengers = num_passengers
        self._total_weight = total_weight
        self._pilot = pilot
        self._crew = crew

    @property
    def total_weight(self):
        """Return the total weight of the flight.

        This is a float value representing the combined weight of the aircraft,
        passengers, cargo, and fuel (in pounds). Must be positive.
        """
        return self._total_weight

    @total_weight.setter
    def total_weight(self, weight):
        """Set the total weight of the flight with validation.

        Validates that the weight is positive and is a float before updating.
        This ensures data integrity for critical flight information.

        Arguments:
            weight (float): The new total weight (must be positive)

        Raises:
            Implicitly fails validation if weight is not positive or not a float
        """
        if weight > 0 and isinstance(weight, float):
            self._total_weight = weight

    @property
    def pilot(self):
        """Return the name of the pilot in command of the flight."""
        return self._pilot

    @pilot.setter
    def pilot(self, new_pilot):
        """Set a new pilot for the flight.

        Arguments:
            new_pilot (str): The name of the new pilot in command
        """
        self._pilot = new_pilot

    @property
    def crew(self):
        """Return the list of crew members assigned to this flight."""
        return self._crew

    @crew.setter
    def crew(self, new_crew):
        """Set the crew members for the flight.

        Arguments:
            new_crew (list): A list of crew member names
        """
        self._crew = new_crew

    def display_flight_data(self):
        """Display all flight information in a formatted table.

        Prints the flight number, origin, destination, number of passengers,
        total weight, pilot name, and crew list in a organized manner.
        """
        print("== Flight ==")
        print("Number:", self.number)
        print("Number of Passengers:", self.num_passengers)
        print("Weight:", self._total_weight)
        print("Pilot:", self._pilot)
        print("Crew:", self._crew)


# Demonstrating the help() function and __doc__ attribute
print("=" * 60)
print("Using help() function:")
print("=" * 60)
help(Flight)

print("\n" + "=" * 60)
print("Using __doc__ attribute:")
print("=" * 60)
print(Flight.__doc__)

print("\n" + "=" * 60)
print("Why help() and __doc__ are helpful:")
print("=" * 60)
explanation = """
1. help() Function:
    - Displays comprehensive documentation including docstrings from the class,
        methods, and properties
    - Shows method signatures and parameters automatically
    - Provides an interactive way to explore code without reading source files
    - Essential for understanding third-party libraries and APIs

2. __doc__ Attribute:
    - Provides direct programmatic access to docstring content
    - Allows automated documentation generation
    - Can be used to dynamically inspect and display documentation at runtime
    - Useful in IDEs and documentation tools to provide inline help

3. Benefits for Development:
    - Reduces need to read source code to understand usage
    - Improves code maintainability through clear documentation
    - Supports IDE features like autocompletion and parameter hints
    - Enables automated testing and validation of documentation
    - Facilitates knowledge transfer between team members
"""
print(explanation)
````
