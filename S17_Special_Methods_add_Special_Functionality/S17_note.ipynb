# Special (Magic) Methods in Python
- Their names start and end with `double underscores`
    - `__methodname__()`

| Special Method | Description               | Example Use Case       |
|----------------|---------------------------|------------------------|
| `__add__()`    | Adds two objects          | `obj1 + obj2`          |
| `__bool__()`   | Returns truth value       | `bool(obj)`            |
| `__str__()`    | Returns string for print  | `print(obj)`           |
| `__eq__()`     | Checks equality           | `obj1 == obj2`         |
| `__repr__()`   | Official string repr.     | `repr(obj)`            |
| `__lt__()`     | Less than comparison      | `obj1 < obj2`          |
| `__len__()`    | Returns length            | `len(obj)`             |
| `__sizeof__()` | Returns memory size       | `obj.__sizeof__()`     |

- Magic: Special Functionality
    - They are meant to be called `indicarectly` through other methods

- __add__():
    - (5).__add__(6)
        - 11
    - (5).__add__(6) == 5+6:
        - True

````python
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Point2D({self.x}, {self.y})"
    
    def __repr__(self):
        return f"Point2D(x={self.x}, y={self.y})"

point = Point2D(3, 4)
print(point)  # Uses __str__
print(repr(point))  # Uses __repr__
````

````python
class Student:
    def __init__(self, name, grade, sudent_id, age):
        self.name = name
        self.grade = grade
        self.student_id = sudent_id  # Unique identifier for the student
        self.age = age  # Age of the student
    
    def __str__(self):
        return f"Student: {self.name}, Grade: {self.grade}, ID: {self.student_id}, Age: {self.age}"
    
    def __repr__(self):
        return f"Student(name='{self.name}', grade={self.grade}, student_id={self.student_id}, age={self.age})"
    
student = Student("Alice", 90, 12345, 20)
print(student)  # Uses __str__
print(repr(student))  # Uses __repr__
````

# __str__() vs .__repr__()
- The special methods __str__() and __repr__() might look similar at first when you read the documentation because they both return a string that describes the object, but they are very different.

- `__str__()`
    - Provides an informal representation of the object meant for the final users.
    - Favores readability over details or precision.
    - Called by the str() , format(), and print() built-in functions.
    - This method differs from object.__repr__() in that there is no expectation that __str__() return a valid Python expression: a more convenient or concise representation can be used.

- `__repr__()`:
    - Provides a formal representation of the object meant for developers.
    - This representation are used for debugging.
    - Called by the repr() built-in function.


````python
class Shirt:
    def __init__(self, color, size, brand):
        self.color = color
        self.size = size
        self.brand = brand
 
    def __str__(self):
        return f"Color: {self.color}; Size: {self.size}; Brand: {self.brand}"
 
    def __repr__(self):
        return f'Shirt("{self.color}", "{self.size}", "{self.brand}")'
 
shirt = Shirt("Blue", "XL", "Retro Land")
 
print("__str__() -", str(shirt))
 
print("\n__repr__() -", repr(shirt))

print("\n__repr__() -", repr(shirt))

# The object can be recreated if needed
new_shirt = eval(repr(shirt))
 
print("\n(New object)", new_shirt)
# Em listas, o Python usa o __repr__ por padrão
inventory = [shirt]
print("List representation:", inventory) 

# Se quiser ver o repr do novo objeto explicitamente:
print("New shirt repr:", repr(new_shirt))
````

# The __repr__ method is essential for debugging as it provides an unambiguous, technical representation of an object. Unlike __str__, it reveals both the data types and the internal state of attributes, allowing developers to catch inconsistencies—such as string vs. integer mismatches—at a glance.
``` python
class FeatureVector:
    def __init__(self, name, values):
        self.name = name
        self.values = values

    def __str__(self):
        """User-friendly summary."""
        return f"Feature: {self.name} (Length: {len(self.values)})"

    def __repr__(self):
        """Technical representation for debugging. 
        The !r flag ensures strings are quoted and types are clear.
        """
        return f"FeatureVector(name={self.name!r}, values={self.values!r})"

# --- DEBUGGING SCENARIO ---

# Object A: Correct list of floats
feature_a = FeatureVector("Price_Index", [10.5, 20.0])

# Object B: Silent Error - numbers passed as strings from a CSV
feature_b = FeatureVector("Price_Index", ["10.5", "20.0"])

print("--- [Output using __str__] ---")
print(f"A -> {feature_a}")
print(f"B -> {feature_b}")
# Observation: They look identical! You can't see the bug here. ✨

print("\n--- [Output using __repr__] ---")
print(f"A -> {repr(feature_a)}")
print(f"B -> {repr(feature_b)}")
# Observation: The quotes around ['10.5', '20.0'] immediately reveal the string bug! 🕵️‍♂️

print("\n--- [List inspection (calls __repr__ automatically)] ---")
print([feature_a, feature_b])
```

````python
class FeatureVector:
    def __init__(self, name, values):
        self.name = name
        self.values = values

    def __str__(self):
        """User-friendly summary."""
        return f"Feature: {self.name} (Length: {len(self.values)})"

    def __repr__(self):
        """Technical representation for debugging. 
        The !r flag ensures strings are quoted and types are clear.
        """
        return f"FeatureVector(name={self.name!r}, values={self.values!r})"

# --- DEBUGGING SCENARIO ---

# Object A: Correct list of floats
feature_a = FeatureVector("Price_Index", [10.5, 20.0])

# Object B: Silent Error - numbers passed as strings from a CSV
feature_b = FeatureVector("Price_Index", ["10.5", "20.0"])

print("--- [Output using __str__] ---")
print(f"A -> {feature_a}")
print(f"B -> {feature_b}")
# Observation: They look identical! You can't see the bug here. ✨

print("\n--- [Output using __repr__] ---")
print(f"A -> {repr(feature_a)}")
print(f"B -> {repr(feature_b)}")
# Observation: The quotes around ['10.5', '20.0'] immediately reveal the string bug! 🕵️‍♂️

print("\n--- [List inspection (calls __repr__ automatically)] ---")
print([feature_a, feature_b])
````

````python
class Backpack:
    def __init__(self):
        self.items = []
    
    def add_item(self, item):
        self.items.append(item)

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
        else:
            print(f"Item '{item}' not found in the backpack.")

    def __len__(self):
        return len(self.items)
    
my_backpack = Backpack("Red", 20)
my_backpack.add_item("Laptop")
my_backpack.add_item("Notebook")
my_backpack.add_item("Pen")
print(f"Backpack has {len(my_backpack)} items.")  # Output: Backpack has 3 items.
print(f"Backpack has {my_backpack.__len__()} items.")  # Output: Backpack has 3 items.
````

# Special Methods: ´__bool__:´


````python
class BankAccount:
    def __init__(self, account_owner, account_number, initial_balance):
        self.account_owner = account_owner
        self.account_number = account_number
        self.initial_balance = initial_balance

    def make_deposit(self, amount):
        self.balance += amount

    def make_withdrawal(self, amount):
        if self.balance - amount >= 0:
            self.balance -= amount
        else:
            print("Insufficient funds for this withdrawal.")

    def __bool__(self):
        # Is the result different from zero ?
        return self.initial_balance > 0

my_account = BankAccount("Alice", "123456789", 1000)
print(my_account.initial_balance)
print(bool(my_account))

my_account.balance = 0

if my_account:
    print("True")
else:
    print("False")


print(bool(my_account))
my_account.make_deposit(500)
````

# __bool__() and __len__() are related?


## `__bool__()` and `__len__()` — a didactic explanation



Em Python, quando você usa `bool(obj)` ou `if obj:`, o interpretador segue esta ordem:



1. **If `__bool__()` exists**, Python uses it.

2. **If `__bool__()` does not exist**, Python tries `__len__()`.

3. If `__len__()` returns `0`, object is **False**; if returns `> 0`, object is **True**.

4. If neither method exists, the object is **True by default**.



That is why empty lists/tuples/strings are falsy: their length is zero.



---



### Example 1: Class with only `__len__()`



```python

class Playlist:

    def __init__(self, songs):

        self.songs = songs



    def __len__(self):

        return len(self.songs)



empty_playlist = Playlist([])

full_playlist = Playlist(["Song A", "Song B"])



print(bool(empty_playlist))  # False (length = 0)

print(bool(full_playlist))   # True  (length > 0)

```



### Example 2: Class with `__bool__()` and `__len__()`



```python

class Wallet:

    def __init__(self, balance):

        self.balance = balance



    def __len__(self):

        return 999  # This will be ignored by bool() in this case



    def __bool__(self):

        return self.balance > 0



empty_wallet = Wallet(0)

rich_wallet = Wallet(100)



print(bool(empty_wallet))  # False (because __bool__ says balance is not > 0)

print(bool(rich_wallet))   # True

```



✅ **Key takeaway:** `__bool__()` has priority over `__len__()`.



### Example 3: Class with neither method



```python

class Student:

    def __init__(self, name):

        self.name = name



s = Student("Marco")

print(bool(s))  # True (default behavior)

```



### Mental model



Think like this:



- `__bool__()` = "Tell me directly if this object is True or False."

- `__len__()` = "If you don't have a direct answer, I will use size as a signal."

- No methods = "I will assume True."


# Rich COmparison Methods:
- Compare objects based on specific criteria.
    - __lt__(): < less than
    - __le__(): <= Less than or Equal to
    - __eq__(): == Equal to
    - __ne__(): != Not equal to
    - __gt__(): > Greater than
    - __ge__(): >=  Greater than or Equal to

````python
print(15 <= 8)
print(4 > 4)
print(5 == 5)
print(6 != 8)
print("Hello" < "World")
print("Hello" == "Hello")
print([1, 2, 3, 4] < [1, 2, 4, 5])
````

````python
class Circle:
    def __init__(self, radius, color):
        self.radius = radius
        self.color = color
    
    def __lt__(self, other):
        return self.radius < other.radius
    
    def __le__(self, other):
        return self.radius <= other.radius

    def __ge__(self, other):
        return (self.radius >= other.radius
                and self.color >= other.color)
    
    def __gt__(self, other):
        return self.radius > other.radius
    
    def __eq__(self, other):
        return (self.radius == other.radius and self.color == other.color)
    
    def __ne__(self, other):
        return (self.radius != other.radius or self.color != other.color)
    
circleA = Circle(5, "Red")
circleB = Circle(7, "Blue")
circleC = Circle(5, "Red")
circleD = Circle(7, "Blue")
print(circleA < circleB)  # True
````

## 🔍 Como funcionam os Rich Comparison Methods?

Os **métodos mágicos de comparação** permitem que você defina **como objetos da sua classe serão comparados** usando operadores como `<`, `>`, `==`, `!=`, `<=`, `>=`.

### 📌 Os 6 métodos principais:

| Método       | Operador | Significado                  |
|--------------|----------|------------------------------|
| `__lt__()`   | `<`      | Less Than (menor que)        |
| `__le__()`   | `<=`     | Less or Equal (menor ou igual) |
| `__eq__()`   | `==`     | Equal (igual)                |
| `__ne__()`   | `!=`     | Not Equal (diferente)        |
| `__gt__()`   | `>`      | Greater Than (maior que)     |
| `__ge__()`   | `>=`     | Greater or Equal (maior ou igual) |

### 🎯 Para que servem?

Sem esses métodos, Python não sabe como comparar objetos da sua classe. Por exemplo:

```python
# Sem __lt__:
circleA < circleB  # ❌ TypeError: '<' not supported

# Com __lt__:
circleA < circleB  # ✅ True (compara os raios)
```

### 🧠 Como funciona internamente?

Quando você escreve `circleA < circleB`, o Python traduz automaticamente para:

```python
circleA.__lt__(circleB)
```

Dentro do método `__lt__`, você define **qual atributo** será usado na comparação.

## 🎓 Resumo Final

### ✅ O que você aprendeu:

1. **Rich Comparison Methods** permitem usar operadores de comparação (`<`, `>`, `==`, etc.) com objetos customizados

2. **Tradução automática**:
   - `obj1 < obj2` → `obj1.__lt__(obj2)`
   - `obj1 == obj2` → `obj1.__eq__(obj2)`
   - E assim por diante...

3. **Você decide o critério**: pode comparar por um atributo, múltiplos atributos, ou qualquer lógica customizada

4. **Aplicação prática**: Permite ordenar listas de objetos com `sorted()`, usar `min()`, `max()`, etc.

### 💡 Dica de boas práticas:

Se você implementar `__eq__()`, deve também implementar `__hash__()` se quiser usar os objetos em sets ou como chaves de dicionários.

### 🚀 Quando usar?

- Quando você precisa **ordenar** objetos customizados
- Quando você quer **comparar** objetos de forma intuitiva
- Quando trabalha com estruturas de dados que exigem comparação (heaps, árvores, etc.)

````python
# 🧪 Exemplo prático: Ordenando círculos

# Criando uma lista de círculos
circles = [
    Circle(10, "Red"),
    Circle(3, "Blue"),
    Circle(7, "Green"),
    Circle(5, "Yellow")
]

print("🔴 Lista original:")
for i, circle in enumerate(circles, 1):
    print(f"   {i}. radius={circle.radius}, color={circle.color}")

# Ordenando pela implementação do __lt__
circles_sorted = sorted(circles)

print("\n✅ Lista ordenada (usando __lt__ - por raio):")
for i, circle in enumerate(circles_sorted, 1):
    print(f"   {i}. radius={circle.radius}, color={circle.color}")

# Comparando dois círculos específicos
print("\n🔍 Comparações específicas:")
small = Circle(3, "Red")
large = Circle(10, "Blue")

print(f"   small < large: {small < large}")
print(f"   small == large: {small == large}")
print(f"   large > small: {large > small}")
````

## 💡 Análise detalhada da classe Circle

Vamos entender **linha por linha** o que está acontecendo:

### 1️⃣ `__lt__(self, other)` - Less Than

```python
def __lt__(self, other):
    return self.radius < other.radius
```

✅ **Compara apenas o raio**  
- `circleA < circleB` → `5 < 7` → `True`

---

### 2️⃣ `__eq__(self, other)` - Equal

```python
def __eq__(self, other):
    return (self.radius == other.radius and self.color == other.color)
```

✅ **Compara raio E cor** (ambos precisam ser iguais)  
- `circleA == circleC` → `(5 == 5 and "Red" == "Red")` → `True`
- `circleA == circleB` → `(5 == 7 and "Red" == "Blue")` → `False`

---

### 3️⃣ `__ge__(self, other)` - Greater or Equal

```python
def __ge__(self, other):
    return (self.radius >= other.radius and self.color >= other.color)
```

⚠️ **Atenção**: Este método tem uma lógica **diferente** dos outros!  
Ele exige que **AMBOS** raio e cor sejam `>=`, o que pode causar resultados inesperados.

**Exemplo problemático:**
```python
circleA = Circle(10, "A")  # radius=10, color="A"
circleB = Circle(5, "Z")   # radius=5, color="Z"

print(circleA >= circleB)  # False! 
# Por quê? Raio: 10 >= 5 ✅, mas Color: "A" >= "Z" ❌
```

📝 **Normalmente**, você deveria usar apenas o raio para manter consistência com `__lt__`.

````python
# 🔬 Testando cada método de comparação

print("=" * 50)
print("🔵 circleA: radius=5, color='Red'")
print("🔵 circleB: radius=7, color='Blue'")
print("🔵 circleC: radius=5, color='Red'")
print("=" * 50)

# __lt__() - Less Than (<)
print("\n1️⃣ __lt__() - Less Than")
print(f"   circleA < circleB: {circleA < circleB}")  # True (5 < 7)
print(f"   circleB < circleA: {circleB < circleA}")  # False (7 < 5)

# __le__() - Less or Equal (<=)
print("\n2️⃣ __le__() - Less or Equal")
print(f"   circleA <= circleB: {circleA <= circleB}")  # True (5 <= 7)
print(f"   circleA <= circleC: {circleA <= circleC}")  # True (5 <= 5)

# __eq__() - Equal (==)
print("\n3️⃣ __eq__() - Equal")
print(f"   circleA == circleC: {circleA == circleC}")  # True (mesmo raio E cor)
print(f"   circleA == circleB: {circleA == circleB}")  # False (raio diferente)

# __ne__() - Not Equal (!=)
print("\n4️⃣ __ne__() - Not Equal")
print(f"   circleA != circleB: {circleA != circleB}")  # True (são diferentes)
print(f"   circleA != circleC: {circleA != circleC}")  # False (são iguais)

# __gt__() - Greater Than (>)
print("\n5️⃣ __gt__() - Greater Than")
print(f"   circleB > circleA: {circleB > circleA}")  # True (7 > 5)
print(f"   circleA > circleB: {circleA > circleB}")  # False (5 > 7)

# __ge__() - Greater or Equal (>=)
print("\n6️⃣ __ge__() - Greater or Equal")
print(f"   circleB >= circleA: {circleB >= circleA}")  # Depende da implementação
print(f"   circleC >= circleA: {circleC >= circleA}")  # Depende da implementação
````
