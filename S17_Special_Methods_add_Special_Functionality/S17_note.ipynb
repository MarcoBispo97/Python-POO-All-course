# Special (Magic) Methods in Python
- Their names start and end with `double underscores`
    - `__methodname__()`

| Special Method | Description               | Example Use Case       |
|----------------|---------------------------|------------------------|
| `__add__()`    | Adds two objects          | `obj1 + obj2`          |
| `__bool__()`   | Returns truth value       | `bool(obj)`            |
| `__str__()`    | Returns string for print  | `print(obj)`           |
| `__eq__()`     | Checks equality           | `obj1 == obj2`         |
| `__repr__()`   | Official string repr.     | `repr(obj)`            |
| `__lt__()`     | Less than comparison      | `obj1 < obj2`          |
| `__len__()`    | Returns length            | `len(obj)`             |
| `__sizeof__()` | Returns memory size       | `obj.__sizeof__()`     |

- Magic: Special Functionality
    - They are meant to be called `indicarectly` through other methods

- __add__():
    - (5).__add__(6)
        - 11
    - (5).__add__(6) == 5+6:
        - True

# Special Methods are also called:
- Magic Methods
- Dunder Methods

The term "dunder" refers to the "double underscore" that is characteristic of the syntax of special methods

It's more conmcise way to refer to the methods without saying "underscore underscore init underscore underscore". We just say "dunder init".

````python
name = 5
print(name)
print(int.__str__(name))
````

````python
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Point2D({self.x}, {self.y})"
    
    def __repr__(self):
        return f"Point2D(x={self.x}, y={self.y})"

point = Point2D(3, 4)
print(point)  # Uses __str__
print(repr(point))  # Uses __repr__
````

````python
class Student:
    def __init__(self, name, grade, sudent_id, age):
        self.name = name
        self.grade = grade
        self.student_id = sudent_id  # Unique identifier for the student
        self.age = age  # Age of the student
    
    def __str__(self):
        return f"Student: {self.name}, Grade: {self.grade}, ID: {self.student_id}, Age: {self.age}"
    
    def __repr__(self):
        return f"Student(name='{self.name}', grade={self.grade}, student_id={self.student_id}, age={self.age})"
    
student = Student("Alice", 90, 12345, 20)
print(student)  # Uses __str__
print(repr(student))  # Uses __repr__
````

# __str__() vs .__repr__()
- The special methods __str__() and __repr__() might look similar at first when you read the documentation because they both return a string that describes the object, but they are very different.

- `__str__()`
    - Provides an informal representation of the object meant for the final users.
    - Favores readability over details or precision.
    - Called by the str() , format(), and print() built-in functions.
    - This method differs from object.__repr__() in that there is no expectation that __str__() return a valid Python expression: a more convenient or concise representation can be used.

- `__repr__()`:
    - Provides a formal representation of the object meant for developers.
    - This representation are used for debugging.
    - Called by the repr() built-in function.

``` python
class Shirt:
    def __init__(self, color, size, brand):
        self.color = color
        self.size = size
        self.brand = brand
 
    def __str__(self):
        return f"Color: {self.color}; Size: {self.size}; Brand: {self.brand}"
 
    def __repr__(self):
        return f'Shirt("{self.color}", "{self.size}", "{self.brand}")'
 
shirt = Shirt("Blue", "XL", "Retro Land")
 
print("__str__() -", str(shirt))
 
print("\n__repr__() -", repr(shirt))
 
# The object can be recreated if needed
new_shirt = eval(repr(shirt))
 
print("\n(New object)", new_shirt)

__str__() - Color: Blue; Size: XL; Brand: Retro Land
 
__repr__() - Shirt("Blue", "XL", "Retro Land")
 
(New object) Color: Blue; Size: XL; Brand: Retro Land
```

````python
class Shirt:
    def __init__(self, color, size, brand):
        self.color = color
        self.size = size
        self.brand = brand
 
    def __str__(self):
        return f"Color: {self.color}; Size: {self.size}; Brand: {self.brand}"
 
    def __repr__(self):
        return f'Shirt("{self.color}", "{self.size}", "{self.brand}")'
 
shirt = Shirt("Blue", "XL", "Retro Land")
 
print("__str__() -", str(shirt))
 
print("\n__repr__() -", repr(shirt))

print("\n__repr__() -", repr(shirt))

# The object can be recreated if needed
new_shirt = eval(repr(shirt))
 
print("\n(New object)", new_shirt)
# Em listas, o Python usa o __repr__ por padrão
inventory = [shirt]
print("List representation:", inventory) 

# Se quiser ver o repr do novo objeto explicitamente:
print("New shirt repr:", repr(new_shirt))
````

# The __repr__ method is essential for debugging as it provides an unambiguous, technical representation of an object. Unlike __str__, it reveals both the data types and the internal state of attributes, allowing developers to catch inconsistencies—such as string vs. integer mismatches—at a glance.
``` python
class FeatureVector:
    def __init__(self, name, values):
        self.name = name
        self.values = values

    def __str__(self):
        """User-friendly summary."""
        return f"Feature: {self.name} (Length: {len(self.values)})"

    def __repr__(self):
        """Technical representation for debugging. 
        The !r flag ensures strings are quoted and types are clear.
        """
        return f"FeatureVector(name={self.name!r}, values={self.values!r})"

# --- DEBUGGING SCENARIO ---

# Object A: Correct list of floats
feature_a = FeatureVector("Price_Index", [10.5, 20.0])

# Object B: Silent Error - numbers passed as strings from a CSV
feature_b = FeatureVector("Price_Index", ["10.5", "20.0"])

print("--- [Output using __str__] ---")
print(f"A -> {feature_a}")
print(f"B -> {feature_b}")
# Observation: They look identical! You can't see the bug here. ✨

print("\n--- [Output using __repr__] ---")
print(f"A -> {repr(feature_a)}")
print(f"B -> {repr(feature_b)}")
# Observation: The quotes around ['10.5', '20.0'] immediately reveal the string bug! 🕵️‍♂️

print("\n--- [List inspection (calls __repr__ automatically)] ---")
print([feature_a, feature_b])
```

````python
class FeatureVector:
    def __init__(self, name, values):
        self.name = name
        self.values = values

    def __str__(self):
        """User-friendly summary."""
        return f"Feature: {self.name} (Length: {len(self.values)})"

    def __repr__(self):
        """Technical representation for debugging. 
        The !r flag ensures strings are quoted and types are clear.
        """
        return f"FeatureVector(name={self.name!r}, values={self.values!r})"

# --- DEBUGGING SCENARIO ---

# Object A: Correct list of floats
feature_a = FeatureVector("Price_Index", [10.5, 20.0])

# Object B: Silent Error - numbers passed as strings from a CSV
feature_b = FeatureVector("Price_Index", ["10.5", "20.0"])

print("--- [Output using __str__] ---")
print(f"A -> {feature_a}")
print(f"B -> {feature_b}")
# Observation: They look identical! You can't see the bug here. ✨

print("\n--- [Output using __repr__] ---")
print(f"A -> {repr(feature_a)}")
print(f"B -> {repr(feature_b)}")
# Observation: The quotes around ['10.5', '20.0'] immediately reveal the string bug! 🕵️‍♂️

print("\n--- [List inspection (calls __repr__ automatically)] ---")
print([feature_a, feature_b])
````

# `__add__()`:
- obj1 + obj2

````python
print(3 + 4)
print((3).__add__(4))

print("Hello "  + "World!")
print(("Hello ").__add__("World!"))

print([1, 2, 3] + [4, 5, 6])
print(([1, 2, 3]).__add__([4, 5, 6]))
````

````python
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other): # (3).__add__(4)
        new_x = self.x + other.x
        new_y = self.y + other.y
        return Point2D(new_x, new_y)
    
    def __str__(self):
        return f"Point2D({self.x}, {self.y})"
    
    def __repr__(self):
        return f"Point2D(x={self.x}, y={self.y})"

pointA = Point2D(1, 2)
pointB = Point2D(3, 4)
PointC = pointA + pointB  # This will call pointA.__add__(pointB)
print(PointC)  # Should print Point2D(4, 6) using __str
print(repr(PointC))  # Should print Point2D(x=4, y=6) using __repr__
````

# `.__getitem__(self, key):`

````python
my_list = [
    "W",
    "o",
    "r",
    "d"
]

print(my_list[0])
print(my_list[1])
print(my_list[2])
print(my_list[3])

print("--- Using __getitem__ directly ---")

print(my_list.__getitem__(0))
print(my_list.__getitem__(1))
print(my_list.__getitem__(2))
print(my_list.__getitem__(3))
````

````python
class Bookshelf:
    def __init__(self):
        self.content = [
            [],
            [],
            []
        ]
    
    def add_book(self, book, location):
        self.content[location].append(book)

    def take_book(self, book, location):
        self.content[location].remove(book)

    def __getitem__(self, location):
        return self.content[location]

my_bookshelf = Bookshelf()

my_bookshelf.add_book("Less Miserables", 0)
my_bookshelf.add_book("The Great Gatsby", 1)
my_bookshelf.add_book("Frankenstein", 2)
print(my_bookshelf[0])  # Should print ['Less Miserables']
print(my_bookshelf[1])  # Should print ['The Great Gatsby']
print(my_bookshelf[2])  # Should print ['Frankenstein']
````

# `.__bool__()`:


````python
class BankAccount:
    def __init__(self, account_owner, account_number, initial_balance=0):
        self.account_owner = account_owner
        self.account_number = account_number
        self.balance = initial_balance

    def make_deposit(self, amount):
        self.balance += amount

    def make_withdrawal(self, amount):
        if self.balance - amount >= 0:
            self.balance -= amount
        else:
            print("Insufficent funds for this withdrawal."
                f"Current balance: {self.balance}, Withdrawal amount: {amount}")

my_account = BankAccount("Alice", "12345", 1000)
print(f"Initial balance: {my_account.balance}")
print(bool(my_account))  # By default, this will print True since the object exists
````
